```python
import requests as req      #Importing Library
import pandas as pd

class Data:

  def __init__(self, url):
    self.url = url


  def download_data(self):      # Downloads json data from url
    res = req.get(self.url)
    return res.json()

  def process_data(self):       # Converts json data to dataframe
    data = self.download_data()
    df = pd.json_normalize(data["pokemon"])
    return df


url = "https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json"      

ex = Data(url)

df = ex.process_data()
df
```

```python
df.info()
```

```python
# Get all Pokemons whose spawn rate is less than 5%
df[df['spawn_chance'] < 0.05]
```

```python
def weakness(l):
  if len(l) < 4:
    return True
  else:
    return False
```

```python
# Get all Pokemons that have less than 4 weaknesses
df[df['weaknesses'].apply(weakness)]
```

```python
# Get all Pokemons that have no multipliers at all
df[df['multipliers'].isnull()]
```

```python
def evol(l):
  try: 
    if len(l)<2:
      return True
    else:
      return False
  except:
    return False
```

```python
# Get all Pokemons that do not have more than 2 evolutions
df[df['next_evolution'].apply(evol)]
```

```python
def time(st):
  if st == 'N/A':
    return "NaN"
  else:
    return st
```

```python
# Get all Pokemons whose spawn time is less than 300 seconds.
data = df.copy()
data['spawn_time'] = data['spawn_time'].apply(time)
data['spawn_time'] = pd.to_datetime(data['spawn_time'])
data['spawn_min'] = data['spawn_time'].dt.hour
data['spawn_sec'] = data['spawn_time'].dt.minute
data['total_sec'] = (data['spawn_min']*60) + data['spawn_sec'] 
data[data['total_sec']<300]
```

```python
def checklen(l):
  if len(l) > 2:
    return True
  else:
    return False
```

```python
# Get all Pokemon who have more than two types of capabilities
df[df['type'].apply(checklen)]
```
