```python
def freq(string):
  arr = string.split(" ")
  d = {}
  for i in arr:
    if i not in d:
      d[i] = 1
    else:
      d[i] += 1

  v = list(d.values())
  m = max(v)

  words = [i for i in d.keys() if d[i]==m]
  word = max(words, key=len)

  return len(word)


st = "write write write all the number from from from 1 to 100"  
st1 = "Hi hi My name is xys xys xys abd aaa and I am learning python"
st2 = "I want want to to be a data scientist scientist"
print(freq(st), " For default given string")
print(freq(st1), " for 1st string")
print(freq(st2), " for 2nd string")
```

Case 1: string = "Hi hi My name is xys xys xys abd aaa and I am learning python" and output is 3 as 'xys' has the highest frequency.

Case 2: string = "I want want to to be a data scientist scientist" and output is 9 as this string 3 words 'want', 'to' and 'scientist' repeat maximum times and so printing the length of word with maximum length here scientist.
