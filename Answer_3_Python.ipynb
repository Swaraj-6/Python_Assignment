```python
import requests as req      #Importing Library
import pandas as pd

class Excel:

  def __init__(self, url, save_path):
    self.url = url
    self.save_path = save_path


  def download_data(self):      # Downloads json data from url
    res = req.get(self.url)
    return res.json()

  def process_data(self):       # Converts json data to dataframe
    data = self.download_data()
    df = pd.json_normalize(data["pokemon"])
    return df

  def save_file(self):        # Converts dataframe to excel and saves it on given location
    df = self.process_data()
    df.to_excel(self.save_path, index=True)
    print("File saved sucessfully")


url = "https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json"     
save_path = "datafile.xlsx"  

ex = Excel(url, save_path)

ex.save_file()

```
