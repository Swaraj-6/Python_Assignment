```python
from collections import Counter
def valid(st):
  dic = {}
  for i in range(len(st)):
    if st[i] not in dic:
      dic[st[i]] = 1
    else:
      dic[st[i]] += 1
  
  ele = dic[st[0]]
  for i in dic.keys():
    if ele != dic[i]:
      return "NO"
    
  return "YES"
  



print(valid("abc"))    #Test case 1
print(valid("Hello") )   #Test case 2
print(valid("xyzusxzysu"))    #Test case 3
```

Case 1: Input string is "abc" where all characters appearing equal number of time so it is valid string.

Case 2: Input string is "Hello" where "l" is repeating itself so it is not a valid string.

Case 3: Input string is "xyzusxzysu" where all characters appearing equal number of time so it is valid string.
